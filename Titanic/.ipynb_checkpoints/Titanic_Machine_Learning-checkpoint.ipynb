# Part 2: Data Cleaning
## This section will clean the train set

```python
#remove passengers with fares that will skew results
train = train[train['Fare']<500]

#remove passengers that never embarked
train = train[~train['Embarked'].isnull()]

#drop duplicate ticket numbers and keep whoever appears first for the ticket number in dataframe
train = train.drop_duplicates(subset=['Ticket'], keep='first')

#drop null ages
train = train[train['Age']!=-5]

#reset index
train = train.reset_index(drop=True)
```

```python
#extract salutations
#this will allow us to go deeper than gender, we can see martitual status as well

train['Prefix']=train['Name'].str.extract('([A-Za-z]+)\.')

#fix some prefixes. all are correct, but some mean the same thing

train.replace({'Prefix' : {'Mme':'Miss', 'Mlle':'Miss', 'Lady':'Mrs', 'Ms':'Miss'}},inplace=True)
```

```python
# convert to numerical categories

cats = ['Name', 'Ticket', 'Cabin', 'Embarked', 'Sex', 'Age', 'Pclass', 'Prefix']

for x in cats:
    train[x] = train[x].astype('category').cat.codes
```

```python
#heatmap of correllation
#because i categorized as numbers, some show a negative correllation which would mean there is an inverse relationship
#this should be ignored in some cases, for example, Fare and Pclass show a -0.6 correllation, when in reality, we
#know they are correllated

plt.rcParams['figure.figsize'] = 20,10
cor = train.corr()
mask = np.zeros_like(cor)
mask[np.triu_indices_from(mask)] = True


x = sns.heatmap(cor, annot=True, mask=mask)
plt.setp(x.get_yticklabels(), rotation=360)
plt.show()
```

# Part 4: Machine Learning - Train and Test
## Predict if a passenger lives or dies

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
#what is the size of our dataset? 
train.shape[0]
```

```python
#I want to take 75% of the train dataset to truly train and the other 25% will serve as a test with the outcomes that 
#are known
532*.75
```

```python
#split the original test into a test and train (test set provided does not have actual outcomes)

test = train[400:]
train = train[:399]
```

```python
#create variables for our target and our variables

train_true = train['Survived']
train_vars = train.columns[2:14]
```

```python
#set our RFC to run through the data twice to create a model

clf = RandomForestClassifier(n_jobs=3)
clf.fit(train[train_vars], train_true)
```

```python
#my model is scoring very high, it could potentially be overfitting

clf.score(train[train_vars], train_true)
```

```python
#my model was able to predict the correct answer 84% of the time
#i no longer believe it was a case of overfitting
#now that the model has been validated, I will apply it to the 'real' dataset

predictions = clf.predict(test[train_vars])
pd.crosstab(test['Survived'], predictions, colnames=['Predicted'])
```
